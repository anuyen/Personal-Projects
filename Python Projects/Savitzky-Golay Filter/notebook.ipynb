```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

## Preparing signal data


```python
x = np.arange(-2,2,0.02)
y = np.array(x**2 + 2*np.sin(x*np.pi))
y = y + np.array(np.random.random(len(x))*2.3)

plt.figure(figsize=(12,9))
plt.plot(x,y,label='y_signal')
plt.legend()
plt.grid(True)
plt.show()
```

Our goal is the 'smooth' above signal using the savgol_filter() function

## Smoothing with Savitzky-Golat filter

To smooth the signal data, Savitzky-Golat filter calculates a polynomial fit of each window based on polynomial degree and window size

    Savitzky-Golat filter function requires:
    _One dimensional array data
    _Window length
    _Polynomial order
    _Other params

```python
y_smooth = signal.savgol_filter(y,window_length=11,polyorder=3,mode='nearest')

plt.figure(figsize=(9, 7))
plt.plot(x, y, label="y_signal")
plt.plot(x, y_smooth, linewidth=3, label="y_smoothed")
plt.legend()
plt.grid(True)
plt.show()
```

Window size and polynomial order parameters can be changed to better smooth the data

```python
ig, ax = plt.subplots(4,figsize=(8,14))
i = 0

# Trying for window size 5, 11, 21, 31
for w_size in [5,11,21,31]:
    y_fit=signal.savgol_filter(y,w_size,3,mode="nearest")
    ax[i].plot(x,y,label="y_sinal",color="blue")
    ax[i].plot(x,y_fit,label='y_smooothed',color='red')
    ax[i].set_title("Window size: " + str(w_size))
    ax[i].legend()
    ax[i].grid(True)
    i+=1

plt.tight_layout()
plt.show()
```
