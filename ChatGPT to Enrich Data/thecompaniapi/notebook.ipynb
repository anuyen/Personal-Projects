```python
import pandas as pd

test_df = pd.read_csv('../output/cleaned_data.csv')

test_df.keys()
```

```python
def fill_domain(df: pd.DataFrame)-> pd.DataFrame:
    """
    Sometimes, domain is missing data or has corrupt domains
    This can be addressed by replacing the bad domain row with 
    data from the company_name_cleaned column

    The company API will also takes straight up company names

    In:  pd.DataFrame[['website_cleaned','company_name_cleaned']]
    Out: pd.DataFrame[['website_cleaned','company_name_cleaned','website_filled']]
    """
    # fill na
    # df = df.insert(0,"website_filled",df.loc[:,'website_cleaned'].fillna(df.loc[:,'company_name_cleaned']))
    df.loc[df['website_cleaned'].isna(), 'website_filled'] = df['company_name_cleaned'].apply(lambda x: str(x).replace(" ",""))

    # fill short
    # this is the official best practice for not getting SettingWithCopy warning
    df.loc[df['website_filled'].str.len() < 4, 'website_filled'] = df['company_name_cleaned'].apply(lambda x: str(x).replace(" ",""))

    return df

df = fill_domain(test_df)

website_df = df[['website_filled']].drop_duplicates(subset=['website_filled']).dropna(subset=['website_filled'])

sample_df = website_df.sample(n=10)
```

```python
import Levenshtein

"""
Levenshtein distance between two strings is minimum number of single-character edits
required to change one more into another
"""
Levenshtein.distance("General Motors","General Motors") # = 0
Levenshtein.distance("general-motors","General Motors") # = 3
Levenshtein.distance("red-hat","Red Hat Inc.") # = 3
# Levenshtein.distance("los-angeles-county","Los Angeles County") # = 5
```

```python
def compare_string(company_name: str, linked_id: str, allowable_coefficient: int)-> bool:
    """
    Levenshtein distance between two strings is minimum number of single-character edits
    required to change one more into another
    """
    return Levenshtein.distance(company_name,linked_id) - len(linked_id.split('-')) < allowable_coefficient
    
```

```python
compare_string("red-hat","Red Hat", 3)
```
